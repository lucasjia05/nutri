# Analyzing results

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
carb_base_1 = pd.read_json("/data/lucasjia/projects/nutri/results/multi-nutrient/sub1/samples_carb_base_20250719_063854.jsonl", lines=True)
carb_base_2 = pd.read_json("/data/lucasjia/projects/nutri/results/multi-nutrient/sub1/samples_carb_base_20250720_033502.jsonl", lines=True)
pred_corr = carb_base_1["pred"].corr(carb_base_2["pred"])
print(pred_corr)
mae = (carb_base_1["pred"] - carb_base_2["pred"]).abs().max()
print(mae)

plt.scatter(x=carb_base_1["pred"], y=carb_base_2["pred"])
plt.hist(x=mae)
```
